# Learning a qaudratic with a simple neural net

Here we want to train a simple neural net to learn the function $f(x)=x^2$ and explore aspects of the landscape of the related loss function. The network we will train, depicted below, has a single input neuron, a single hidden layer with two neurons and a single output neuron. 

<img style="border:5px" src="Images/net_1_2_1.png" height="500" width="475">

```python
import numpy as np
import neural_network as nn
import matplotlib.pyplot as plt

def func(x):
    return x**2



def loss(net, xs, ys):
    dists = [ (y - net.feed_forward(x))**2
              for x,y in zip(xs,ys) ]
    
    return sum(dists)/len(dists)
```

Generate data:

```python
num_rand_points = 10000

random_xs = 2*np.random.rand(num_rand_points) - 1
target_ys = func(random_xs)
training_data = [ (x,y) for x,y in zip(random_xs,target_ys) ]

num_test_points = 1000

random_xs = 2*np.random.rand(num_test_points) - 1
target_ys = func(random_xs)
test_data = [ (x,y) for x,y in zip(random_xs,target_ys) ]
```

Train a neural net to learn the quadratic function:

```python
net = nn.NeuralNetwork([1,2,1])
x = np.arange(-1,1,0.01)
num = 50
y = np.zeros([num+1,len(x)])

y[0,:] = np.asarray( list(map(net.feed_forward, x)) ).reshape(len(x))

print('Training network')
for i in range(1,num+1):
    
    net.stochastic_gradient_decent(training_data, 2, 50, 1.0, test_data)
    
    y[i,:] = np.asarray( list(map(net.feed_forward, x)) ).reshape(len(x))
print('Finished training')

weights = net.weights
biases = net.biases
```

Plot evolution of net:

```python
%%capture
from matplotlib import animation
from IPython.display import HTML

fig, ax = plt.subplots()

ax.set_xlim([-1,1])
ax.set_ylim([0,1])
ax.set_xlabel('x')
ax.set_ylabel('y')

dots, = ax.plot([],[],'o',markerSize=1)
line, = ax.plot([],[],lw=4, c='darkslategray')

def init():
    dots.set_data(random_xs ,target_ys)
    return (dots,)

def animate(i):
    line.set_data(x, y[i,:])
    return (line,)

anim = animation.FuncAnimation(fig, animate, init_func = init, frames=num+1, interval=30, blit=True)
```

Evaluate change in loss due to a variation in one of the networks parameters:

```python
from itertools import chain

c = np.zeros( [sum([w.size for w in chain(weights,biases)]) ,len(x)])

k = 0
for param in chain(weights,biases):
    
    n,m = param.shape
    
    for i in range(n):
        for j in range(m):
            
            omega = param[i,j]

            for l,xi in enumerate(x):
                param[i,j] = omega + 10*xi
                c[k,l] = loss(net, random_xs, target_ys)
            
            print('calculated change in loss due to varying omega_{0}'
                  .format(k) )
            param[i,j] = omega
            k += 1
```

Plot the graphs of the above changes:

```python
plt.figure(2)
for pic in range(c.shape[0]):
    plt.plot(10*x, c[pic,:], label='{0}'.format(pic) )

plt.title(r'Loss of net in vicinity of trained weigths $\omega_i$')
plt.xlabel(r'$\delta \omega_i$')
plt.ylabel(r'Loss$(\omega +\delta \omega_i)$')
plt.legend()
```
